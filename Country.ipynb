```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

shp1 = os.path.dirname('/home/masum/Documents/heyBro/31-05/ne_10m_admin_0_countries/ne_10m_admin_0_countries.shp')
shp2 = os.path.dirname('/home/masum/Documents/heyBro/CHN_ADM1/CHN_ADM1.shp')

countries = gpd.read_file(shp1)
states = gpd.read_file(shp2)

China = countries[countries['NAME'] == 'China']
Taiwan = countries[countries['NAME'] == 'Taiwan']


Shaanxi = states[states['NAME'] == 'Shaanxi']
Ningxia = states[states['NAME'] == 'Ningxia Hui']

china_boundary_map = states.boundary.plot(figsize=(18, 12), color='Black', linewidth=1)

China.plot(ax=china_boundary_map,  color="#cdc7eb", alpha=0.5)
Taiwan.plot(ax=china_boundary_map,  color="#cdc7eb")
Shaanxi.plot(ax=china_boundary_map,  color="PaleTurquoise")
Ningxia.plot(ax=china_boundary_map,  color="#fff44f", alpha=0.5)
plt.axis('off')
plt.savefig('/home/masum/Documents/heyBro/31-05/China.png', bbox_inches='tight', dpi=300)
```
