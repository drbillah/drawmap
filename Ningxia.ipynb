```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

shp1 = os.path.dirname('/home/masum/Documents/heyBro/CHN_ADM1/CHN_ADM1.shp')
shp2 = os.path.dirname('/home/masum/Documents/heyBro/CHN_ADM3/CHN_ADM3.shp')

states = gpd.read_file(shp1)
district = gpd.read_file(shp2)


#Shaanxi = states[states['NAME'] == 'Shaanxi']
Ningxia = states[states['NAME'] == 'Ningxia Hui']

Pingluo = district[district['NAME'] == 'Pingluo County']
Tongxin = district[district['NAME'] == 'Tongxin County']
Xiji = district[district['NAME'] == 'Xiji County']

ningxia_boundary_map = Ningxia.boundary.plot(figsize=(18, 12), color='Black', linewidth=1)

Ningxia.plot(ax=ningxia_boundary_map,  color="#fff44f", alpha=0.5)
Pingluo.plot(ax=ningxia_boundary_map,  color="MistyRose")
Tongxin.plot(ax=ningxia_boundary_map,  color="PaleTurquoise")
Xiji.plot(ax=ningxia_boundary_map,  color="Plum")

plt.axis('off')
plt.savefig('/home/masum/Documents/heyBro/31-05/Ningxia.png', bbox_inches='tight', dpi=300)
```
